Experiment 1 ist einfach geimpft, da heißen die Gruppen im Objekt: aaUntr, adeno, attenuated, mRNA – Adeno ist ein selbsgemachter Adenovirusimpfstoff, mRNA ist Biontech, attenuated ist der live-attenuated virus um den es hier vor allem geht, der wird in die Nase gegeben.

Die Zellen wurden in vier Pools sequenziert, und jeder Pool wurde in vier libraries (die dann part1-4 heißen) aufgearbeitet, deren Daten man dann natürlich mergen kann. Das ist die Zuordnung der CMOs (A=attenuated, B=adeno, C=mRNA, D= aaUntr; LU ist Lunge, BL ist Blut, NM ist nasal mucosa):

Beim zweiten Experiment wurden zweifach geimpft, hier gibt es noch die Kombination mRNA+attenuated. aaUntr (X), adeno2x (B), att2x (A), mRNA2x (C), mRNAatt (D)

```python
spath = '/fast/scratch/users/peidlis_c/CovHamster/Impf/'
libs = [x for x in get_subfolders(spath, False) if 'Pool' in x]
libs
```

```python
dict_sample_type = {
    'LU' : 'lung',
    'BL' : 'blood',
    'NM' : 'nasal mucosa',
}

dict_condition_exp1 = {
    'A': 'attenuated',
    'B': 'adeno',
    'C': 'mRNA',
    'D': 'untreated'
}

dict_condition_exp2 = {
    'X': 'untreated',
    'B': 'adeno+adeno',
    'A': 'attenuated+attenuated',
    'C': 'mRNA+mRNA',
    'D': 'mRNA+attenuated'
}

edict_exp1 = {
    'A': 'att', 
    'B': 'adeno', 
    'C': 'mRNA', 
    'D': 'aaUntr'
}

edict_exp2 = {
    'X': 'aaUntr', 
    'B': 'adeno2x', 
    'A': 'att2x', 
    'C': 'mRNA2x', 
    'D': 'mRNAatt'
}
```

```python
adatas = {}
for lib in libs:
    if 'outs' in get_subfolders(f'{spath}/{lib}', False):
        print(lib)
        samples = get_subfolders(f'{spath}/{lib}/outs/per_sample_outs/', False)
        for sample in samples:
            tdata = sc.read_10x_h5(f'{spath}/{lib}/outs/per_sample_outs/{sample}/count/sample_feature_bc_matrix.h5')
            
            tdata.var_names_make_unique()
            tdata.obs_names = [x.replace('-1', '') for x in tdata.obs_names]
            
            clear_output()
            
            experiment, pool = lib.split('_')
            tdata.obs['experiment'] = experiment
            tdata.obs['pool'] = pool
            tdata.obs['library'] = lib
            tdata.obs['sample_type'] = dict_sample_type[sample.split('_')[0]]
            cdict = eval(f'dict_condition_{experiment}')
            edict = eval(f'edict_{experiment}')
            tdata.obs['condition'] = cdict[sample[-2]]
            tdata.obs['sample_number'] = sample[-1]
            sample_id = f'{pool}_{sample}'
            tdata.obs['sample_id'] = sample_id
            
            tdata.obs['cell_id'] = [f'{edict[sample[-2]]}_{pool[-1]}_{sample[-1]}_{x}-1' for x in tdata.obs_names]
            
            adatas[sample_id] = tdata
adata = sc.concat(adatas, index_unique='-')
print('Loaded libs:', pd.unique(adata.obs.library))
adata
```
